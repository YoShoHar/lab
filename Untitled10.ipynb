**Check data**

```python
#check data.info
print(df_customers.info())
print("="*100)
print(df_transaction.info())
print(f'[{"*"*100}]')
#check missing data
print(df_customers.isnull().sum())
print("="*100)
print(df_transaction.isnull().sum())
print(f'[{"*"*100}]')
#check describe
print(df_customers.describe())
print("="*100)
print(df_transaction.describe())
print(f'[{"*"*100}]')
```

**Sale by Age**

```python
print(sns.__version__)

plt.figure(figsize=(15,5))

kde = sns.kdeplot(df_customers["age"], color="red")
kde_data = kde.get_lines()[0].get_data()
kde.remove()

sns.histplot(df_customers["age"], bins=15, color="#1f77b4", alpha=0.6)

sns.lineplot(x=kde_data[0], y=kde_data[1] * len(df_customers) * 3,
             color="red", linewidth=1.5, ax=plt.gca())

plt.title("Sale by Age", fontsize =14)
plt.xlabel("Age", fontsize =12)
plt.ylabel("Frequency", fontsize =12)
plt.show()
#
```

**Sale by Gender**

```python
plt.figure(figsize=(6,4))
custom_colors = {"Female": "#ff9999", "Male": "#66b3ff", "Other": "#99ff99"}
sns.countplot(
    x=df_customers["gender"],
    palette=[custom_colors[val]for val in df_customers["gender"].unique()]
  )
plt.title("Sale by Gender", fontsize =14)
plt.xlabel("Gender", fontsize =12)
plt.ylabel("Frequency", fontsize =12)
plt.show()
#
```

**Average Transaction Value by Age Group**

```python
#merge CustomerID and transaction
df_merged = df_transaction.merge(df_customers, on ="customer_id")
#merge by age group and mean
df_age_group = df_merged.groupby("age_group")["list_price"].mean().reset_index()

#draw
plt.figure(figsize=(6,4))
sns.barplot(x="age_group", y="list_price", data=df_age_group)
plt.title("Average Transaction Value by Age Group", fontsize =14)
plt.xlabel("Age Group", fontsize =12)
plt.ylabel("Average Transaction Value", fontsize =12)
plt.xticks(rotation=45)
plt.show()
#
```

**Frequency Distributio**

```python
df_freq = df_merged.groupby("customer_id")["transaction_id"].count().reset_index()
df_freq.columns = ["customer_id", "frequency"]

print(df_freq["frequency"].min(), df_freq["frequency"].max())


plt.figure(figsize=(10,5))

kde = sns.kdeplot(df_freq["frequency"], color="red", bw_adjust=1, clip=(1, 14))
kde_data = kde.get_lines()[0].get_data()

bins = np.arange(1, 15, 1)

sns.histplot(df_freq["frequency"], bins=15, color="#1f77b4", alpha=0.8, edgecolor="black")

sns.lineplot(
    x=kde_data[0],
    y=kde_data[1] * len(df_freq["frequency"]),
    color="red",
    linewidth=1.5,
    ax=plt.gca()
)

plt.title("Frequency Distribution", fontsize =14)
plt.xlabel("Frequency", fontsize =12)
plt.ylabel("Customers", fontsize =12)
plt.xticks(np.arange(1, 15, 1))
plt.show()
#
```
