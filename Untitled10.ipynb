**Check data**

```python
#check data.info
print(df_customers.info())
print("="*100)
print(df_transaction.info())
print(f'[{"*"*100}]')
#check missing data
print(df_customers.isnull().sum())
print("="*100)
print(df_transaction.isnull().sum())
print(f'[{"*"*100}]')
#check describe
print(df_customers.describe())
print("="*100)
print(df_transaction.describe())
print(f'[{"*"*100}]')
```

**Sale by Age**

```python
print(sns.__version__)

plt.figure(figsize=(15,5))

kde = sns.kdeplot(df_customers["age"], color="red")
kde_data = kde.get_lines()[0].get_data()
kde.remove()

sns.histplot(df_customers["age"], bins=15, color="#1f77b4", alpha=0.6)

sns.lineplot(x=kde_data[0], y=kde_data[1] * len(df_customers) * 3,
             color="red", linewidth=1.5, ax=plt.gca())

plt.title("Sale by Age", fontsize =14)
plt.xlabel("Age", fontsize =12)
plt.ylabel("Frequency", fontsize =12)
plt.show()
#
```

**Sale by Gender**

```python
plt.figure(figsize=(6,4))
custom_colors = {"Female": "#ff9999", "Male": "#66b3ff", "Other": "#99ff99"}
sns.countplot(
    x=df_customers["gender"],
    palette=[custom_colors[val]for val in df_customers["gender"].unique()]
  )
plt.title("Sale by Gender", fontsize =14)
plt.xlabel("Gender", fontsize =12)
plt.ylabel("Frequency", fontsize =12)
plt.show()
#
```

**Average Transaction Value by Age Group**

```python
#merge CustomerID and transaction
df_merged = df_transaction.merge(df_customers, on ="customer_id")
#merge by age group and mean
df_age_group = df_merged.groupby("age_group")["list_price"].mean().reset_index()

#draw
plt.figure(figsize=(6,4))
sns.barplot(x="age_group", y="list_price", data=df_age_group)
plt.title("Average Transaction Value by Age Group", fontsize =14)
plt.xlabel("Age Group", fontsize =12)
plt.ylabel("Average Transaction Value", fontsize =12)
plt.xticks(rotation=45)
plt.show()
#
```

**Frequency Distributio**

```python
df_freq = df_merged.groupby("customer_id")["transaction_id"].count().reset_index()
df_freq.columns = ["customer_id", "frequency"]

print(df_freq["frequency"].min(), df_freq["frequency"].max())


plt.figure(figsize=(10,5))

kde = sns.kdeplot(df_freq["frequency"], color="red", bw_adjust=1, clip=(1, 14))
kde_data = kde.get_lines()[0].get_data()

bins = np.arange(1, 15, 1)

sns.histplot(df_freq["frequency"], bins=15, color="#1f77b4", alpha=0.8, edgecolor="black")

sns.lineplot(
    x=kde_data[0],
    y=kde_data[1] * len(df_freq["frequency"]),
    color="red",
    linewidth=1.5,
    ax=plt.gca()
)

plt.title("Frequency Distribution", fontsize =14)
plt.xlabel("Frequency", fontsize =12)
plt.ylabel("Customers", fontsize =12)
plt.xticks(np.arange(1, 15, 1))
plt.show()
#
```

**Total Revenue by Month**

```python
df_transaction["transaction_date"] = pd.to_datetime(df_transaction["transaction_date"])

# Create a Month column
df_transaction["Month"] = df_transaction["transaction_date"].dt.to_period("M").astype(str)

# Compute total revenue per month
df_monthly_revenue = df_transaction.groupby("Month")["list_price"].sum().reset_index()

# Plot monthly revenue trend
plt.figure(figsize=(12,5))
sns.lineplot(x="Month", y="list_price", data=df_monthly_revenue, marker="o")
plt.title("Total Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle("Combined Visualizations", fontsize=16)

# Chart 1: Histogram + KDE for Age Distribution
ax1 = axes[0, 0]
kde = sns.kdeplot(df_customers["age"], color="red", ax=ax1)
kde_data = kde.get_lines()[0].get_data()
ax1.clear()

sns.histplot(df_customers["age"], bins=15, color="#1f77b4", alpha=0.6, ax=ax1)
sns.lineplot(x=kde_data[0], y=kde_data[1] * len(df_customers) * 3,
             color="red", linewidth=1.5, ax=ax1)

ax1.set_title("Sales by Age", fontsize=14)
ax1.set_xlabel("Age", fontsize=12)
ax1.set_ylabel("Frequency", fontsize=12)

# Chart 2: Sales by Gender
ax2 = axes[0, 1]
custom_colors = {"Female": "#ff9999", "Male": "#66b3ff", "Other": "#99ff99"}
sns.countplot(
    x=df_customers["gender"],
    palette=[custom_colors[val] for val in df_customers["gender"].unique()],
    ax=ax2
)
ax2.set_title("Sales by Gender", fontsize=14)
ax2.set_xlabel("Gender", fontsize=12)
ax2.set_ylabel("Frequency", fontsize=12)

# Chart 3: Average Transaction Value by Age Group
ax3 = axes[0, 2]
df_merged = df_transaction.merge(df_customers, on="customer_id")
df_age_group = df_merged.groupby("age_group")["list_price"].mean().reset_index()

sns.barplot(x="age_group", y="list_price", data=df_age_group, ax=ax3)
ax3.set_title("Average Transaction Value by Age Group", fontsize=14)
ax3.set_xlabel("Age Group", fontsize=12)
ax3.set_ylabel("Average Transaction Value", fontsize=12)
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=45)

# Chart 4: Customer Frequency Distribution
ax4 = axes[1, 0]
df_freq = df_merged.groupby("customer_id")["transaction_id"].count().reset_index()
df_freq.columns = ["customer_id", "frequency"]

kde = sns.kdeplot(df_freq["frequency"], color="red", bw_adjust=1, clip=(1, 14), ax=ax4)
kde_data = kde.get_lines()[0].get_data()

sns.histplot(df_freq["frequency"], bins=15, color="#1f77b4", alpha=0.8, edgecolor="black", ax=ax4)
sns.lineplot(
    x=kde_data[0],
    y=kde_data[1] * len(df_freq["frequency"]),
    color="red",
    linewidth=1.5,
    ax=ax4
)

ax4.set_title("Customer Frequency Distribution", fontsize=14)
ax4.set_xlabel("Frequency", fontsize=12)
ax4.set_ylabel("Number of Customers", fontsize=12)
ax4.set_xticks(np.arange(1, 15, 1))
ax4.set_xticklabels(np.arange(1, 15, 1))

# Chart 5: Monthly Revenue Trend
ax5 = axes[1, 1]
df_transaction["transaction_date"] = pd.to_datetime(df_transaction["transaction_date"])

# Create a Month column
df_transaction["Month"] = df_transaction["transaction_date"].dt.to_period("M").astype(str)

# Compute total revenue per month
df_monthly_revenue = df_transaction.groupby("Month")["list_price"].sum().reset_index()

# Plot monthly revenue trend
sns.lineplot(x="Month", y="list_price", data=df_monthly_revenue, marker="o", ax=ax5)
ax5.set_title("Total Revenue by Month", fontsize=14)
ax5.set_xlabel("Month", fontsize=12)
ax5.set_ylabel("Revenue", fontsize=12)
ax5.set_xticklabels(ax5.get_xticklabels(), rotation=45)

# Adjust layout for better visualization
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
