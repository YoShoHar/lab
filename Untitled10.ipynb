**Average Transaction Value by Age Group**

```python
#merge CustomerID and transaction
df_merged = df_transaction.merge(df_customers, on ="customer_id")
#merge by age group and mean
df_age_group = df_merged.groupby("age_group")["list_price"].mean().reset_index()
print(df_age_group["age_group"].unique())
age_order = ["Under 30", "30-39", "40-49", "50-59", "60-69", "Above 70"]
#draw
plt.figure(figsize=(6,4))
sns.barplot(x="age_group", y="list_price", data=df_age_group, order=age_order)
plt.title("Average Transaction Value by Age Group", fontsize =14)
plt.xlabel("Age Group", fontsize =12)
plt.ylabel("Average Transaction Value", fontsize =12)
plt.xticks(rotation=45)
plt.show()
#
```

**Frequency Distributio**

```python
df_freq = df_merged.groupby("customer_id")["transaction_id"].count().reset_index()
df_freq.columns = ["customer_id", "frequency"]

print(df_freq["frequency"].min(), df_freq["frequency"].max())


plt.figure(figsize=(10,5))

kde = sns.kdeplot(df_freq["frequency"], color="red", bw_adjust=1, clip=(1, 14))
kde_data = kde.get_lines()[0].get_data()

bins = np.arange(1, 15, 1)

sns.histplot(df_freq["frequency"], bins=15, color="#1f77b4", alpha=0.8, edgecolor="black")

sns.lineplot(
    x=kde_data[0],
    y=kde_data[1] * len(df_freq["frequency"]),
    color="red",
    linewidth=1.5,
    ax=plt.gca()
)

plt.title("Frequency Distribution", fontsize =14)
plt.xlabel("Frequency", fontsize =12)
plt.ylabel("Customers", fontsize =12)
plt.xticks(np.arange(1, 15, 1))
plt.show()
#
```

**Total Revenue by Month**

```python
df_transaction["transaction_date"] = pd.to_datetime(df_transaction["transaction_date"])

# Create a Month column
df_transaction["Month"] = df_transaction["transaction_date"].dt.to_period("M").astype(str)

# Compute total revenue per month
df_monthly_revenue = df_transaction.groupby("Month")["list_price"].sum().reset_index()

# Plot monthly revenue trend
plt.figure(figsize=(12,5))
sns.lineplot(x="Month", y="list_price", data=df_monthly_revenue, marker="o")
plt.title("Total Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

**Top10 Product**

```python
# Calculate total sales per product
df_transaction["quantity_sold"] = df_transaction.groupby(["brand", "product_line"])["transaction_id"].transform("count")

df_transaction["total_sales"] = df_transaction["list_price"] * df_transaction["quantity_sold"]
top_products = df_transaction.groupby(["brand", "product_line"])["total_sales"].sum().reset_index()

# Sort and get top 10 products
top_products = top_products.sort_values(by="total_sales", ascending=True).head(10)

# Format labels
top_products["label"] = top_products["product_line"] + " " + top_products["brand"]

# Create bar chart
fig = go.Figure()

fig.add_trace(go.Bar(
    y=top_products["label"],
    x=top_products["total_sales"],
    orientation='h',
    marker=dict(color='blue'),
    name="Total Sales"
))

# Update layout
fig.update_layout(
    title="Top 10 Products by Total Sales",
    xaxis_title="Total Sales ($)",
    yaxis_title="Products",
    height=600
)

# Show figure
fig.show()
```

**Customer base**

```python
# Read state data from GeoJSON
geo_data = gpd.read_file('/content/australian-states.json')

# Read postcode data from JSON
df_postcodes = pd.read_json("australian_postcodes.json")
display(df_postcodes)

# Merge state and postcode data
customer_base = pd.merge(df_transaction, df_address, on="customer_id")

# Count the number of customers per state
state_counts = customer_base.groupby("state").size().reset_index(name="customer_count")
display(state_counts)

# Get the list of states from GeoJSON
all_states = geo_data[["STATE_NAME"]].drop_duplicates()

# Merge customer count with the list of states
merged_data = all_states.merge(
    state_counts, left_on="STATE_NAME", right_on="state", how="left"
).fillna(0)

# Convert GeoDataFrame to GeoJSON
geojson_data = geo_data.__geo_interface__

# Draw a choropleth map for the states
fig2 = px.choropleth(
    merged_data,
    geojson=geojson_data,
    locations="STATE_NAME",
    featureidkey="properties.STATE_NAME",
    color="customer_count",
    color_continuous_scale="Blues",
    title="Customer Distribution by State & Postcode in Australia"
)

# Add scatter plot for postcodes
fig2.add_trace(
    go.Scattergeo(
        lon=df_postcodes["long"],
        lat=df_postcodes["lat"],
        mode="markers",
        marker=dict(size=4, color="red"),
        name="Postcodes"
    )
)

# Update the map to fit both choropleth and scatter plots
fig2.update_geos(
    projection_scale=6,  # Adjust zoom level (increase value for closer zoom)
    center={"lat": -25, "lon": 135},  # Center the map on Australia
    showframe=False,
    showcoastlines=True
)

fig2.update_layout(
    width=1800,  # Increase width
    height=1000   # Increase height
)

fig2.show()
```
